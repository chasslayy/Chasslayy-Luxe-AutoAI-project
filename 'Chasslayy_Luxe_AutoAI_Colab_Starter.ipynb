````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

X = df.drop(columns=["purchase_made"])
y = df["purchase_made"]

cat_cols = ["engagement_level","product_category_viewed","avg_order_value_bucket","traffic_source"]
num_cols = [c for c in X.columns if c not in cat_cols]

preprocess = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
    ("num", "passthrough", num_cols)
])

clf = Pipeline([
    ("prep", preprocess),
    ("model", LogisticRegression(max_iter=1000))
])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
clf.fit(X_train, y_train)
pred = clf.predict(X_test)

print("Accuracy:", round(accuracy_score(y_test, pred), 4))
print("Confusion matrix:\n", confusion_matrix(y_test, pred))
````

## Re-train Model with Class Weights

### Subtask:
Re-train the Logistic Regression model using `class_weight='balanced'` to inherently handle the class imbalance present in the 'purchase_made' column.


**Reasoning**:
To address the class imbalance, I will re-train the Logistic Regression model by instantiating a new model with `class_weight='balanced'` and `max_iter=1000`, creating a new pipeline with this model, fitting it to the training data, and then making predictions on the test set.



````python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# 1. Instantiate a new LogisticRegression model with class_weight='balanced'
model_balanced = LogisticRegression(class_weight='balanced', max_iter=1000)

# 2. Create a Pipeline named clf_balanced
clf_balanced = Pipeline([
    ("prep", preprocess),
    ("model", model_balanced)
])

# 3. Fit the clf_balanced pipeline to the training data
clf_balanced.fit(X_train, y_train)

# 4. Use the clf_balanced pipeline to make predictions on the test data
pred_balanced = clf_balanced.predict(X_test)

print("Balanced model predictions generated.")
````

**Reasoning**:
Now that the balanced model has been trained and predictions generated, I need to evaluate its performance using various metrics: accuracy, confusion matrix, precision, recall, and F1-score.
````python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

# Evaluate the balanced model
accuracy_balanced = accuracy_score(y_test, pred_balanced)
conf_matrix_balanced = confusion_matrix(y_test, pred_balanced)
precision_balanced = precision_score(y_test, pred_balanced)
recall_balanced = recall_score(y_test, pred_balanced)
f1_balanced = f1_score(y_test, pred_balanced)

print("Balanced Model Evaluation:")
print(f"Accuracy: {accuracy_balanced:.4f}")
print(f"Confusion Matrix:\n{conf_matrix_balanced}")
print(f"Precision: {precision_balanced:.4f}")
print(f"Recall: {recall_balanced:.4f}")
print(f"F1-Score: {f1_balanced:.4f}")
````

````python
# STEP 1: Create a privacy-safe estimated order value
bucket_base = df["avg_order_value_bucket"].map({
    "low": 35,
    "medium": 75,
    "high": 140
}).astype(float)

category_boost = df["product_category_viewed"].map({
    "digital": -10,
    "beauty": 0,
    "wig": 40
}).astype(float)

noise = np.random.normal(loc=0, scale=12, size=len(df))

df["order_value_est"] = (bucket_base + category_boost + noise).clip(10, 300).round(2)

df[["avg_order_value_bucket", "product_category_viewed", "order_value_est"]].head()
````

````python
# STEP 2: Create purchase_made target
base = 0.08
p = (
    base
    + 0.03 * (df["sessions_last_30d"] / 10)
    + 0.08 * (df["prior_purchases"] / 6)
    + 0.10 * df["engagement_level"].map({"low": 0.0, "medium": 0.5, "high": 1.0})
    - 0.06 * (df["days_since_last_interaction"] / 60)
    + 0.04 * df["discount_used"]
    + 0.03 * df["traffic_source"].isin(["instagram", "tiktok"]).astype(int)
).clip(0.01, 0.95)

df["purchase_made"] = np.random.binomial(1, p)

df["purchase_made"].value_counts(normalize=True)
````

````python
# STEP 3: Create high_value_purchase target
HIGH_VALUE_THRESHOLD = 120

df["high_value_purchase"] = (
    (df["purchase_made"] == 1) & (df["order_value_est"] >= HIGH_VALUE_THRESHOLD)
).astype(int)

df[["purchase_made", "order_value_est", "high_value_purchase"]].head(10)
````
